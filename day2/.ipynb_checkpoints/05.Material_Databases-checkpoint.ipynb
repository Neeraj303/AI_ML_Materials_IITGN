# Materials Project



The Materials Project is an open-access platform and database that provides materials scientists,        researchers, and engineers with a wealth of information related to materials properties. ### Using the requests package to get data

Having looked at the API, we now will try to download data with elements 'Si', 'O' using the requests module.

```python
import requests
import warnings
warnings.filterwarnings('ignore')

url = "https://api.materialsproject.org/materials/summary/"
kwargs = {
    'headers': {"X-API-KEY": mp_api_key},
    'params': {
        # see https://api.materialsproject.org/docs#/Materials%20Summary
        'formula': 'SiO2',
        '_fields': ','.join(
            [
                'material_id',
                'formula_pretty',
                'density',
                'structure',
                'symmetry',
                'band_gap'
            ]
        ),
    },
}

response = requests.get(url, **kwargs)
```

```python
response.text
```

### Understanding the Data structure of response from HTTP request

```python
import json

query = response.json()
with open('hi.json', 'w') as json_file:
    json.dump(query, json_file)
```

```python
query.keys()
```

##### Turn the nested-dictionary (response.json()) in to a pandas dataframe


```python
import pandas as pd
df = pd.DataFrame(query['data'])
#df
```

```python
df
```

```python
# Load the Python Package #
from mp_api.client import MPRester

"""
Instantiate the MPRester Object that takes in your API key helping to establish a connection between so that API calls can
be made through mpr object 
"""
mpr = MPRester(mp_api_key)
```

```python
mpr.materials.available_fields
```

```python
# For a specific end-point, can we find the available fields to query #
mpr.materials.available_fields
```

One could instantiate the MPRester class as shown above but the following approach closes the client-server connection post the execution of code within the 'with' statement. Useful if you are making multiple api calls. Analogous to file operations.

Let us try and use it to find the available fields in the elasticity endpoint. 


```python
print(f"Instantiate MPRester - {mpr.materials.elasticity.available_fields}")

with MPRester(mp_api_key) as mpr:
    print(f"Using 'with' Statement- {mpr.materials.elasticity.available_fields}")
```

Great! We've learned how to make API calls using the MPRester package. Now, let's explore more techniques to utilise the query module. But before we do that, let's familiarize ourselves with some methods of MPRester. As we've observed on the website, all the data is linked to a single entry known as the **material ID**, so lets begin with it.

#### Query Material id for a desired chemical formula

```python
with MPRester(mp_api_key) as mpr:
    query = mpr.get_materials_ids("SiO2")
```

```python
query[:5]
```

```python
y  = [x.string for x in query]
```

```python
with MPRester(mp_api_key) as mpr:
    print(mpr.get_database_version())
```

#### Query Structure for a material id

```python
with MPRester(mp_api_key) as mpr:
    struct = mpr.get_structure_by_material_id('mp-6922')
    
struct
```

```python
struct
```

#### Query all the data for a chemical formula

```python
with MPRester(mp_api_key) as mpr:
    out = mpr.materials.search(formula='TiO2')
print(len(out))
```

```python
out[0].dict().keys()
```

```python
mpr.materials.thermo
```

#### Specifically query the thermodynamic data for a chemical formula

```python
with MPRester(mp_api_key) as mpr:
    out = mpr.materials.thermo.search(formula='TiO2')
out[0].dict().keys()
```

#### Query for a list of material ids

```python
# Pass Known material ids #
with MPRester(mp_api_key) as mpr:
    x = mpr.materials.search(material_ids = y)
```

```python
x[0].dict()
```

#### Query certain elemental composition with specific property list

```python
with MPRester(mp_api_key) as mpr:
    x = mpr.materials.search(formula="**O3", fields = ['materials_id', "structure", "density", "formula_pretty", 
                                                       "elasticity" ])
x[0]
```

#### Query EOS for a given material property

**Note**: We changed the end point from 'materials' to 'eos'

```python
with MPRester(mp_api_key) as mpr:
    eos = mpr.eos.get_data_by_id("mp-149")


import matplotlib.pyplot as plt

fig = plt.figure(figsize =(4, 3),dpi=150)
plt.scatter(eos.dict()['volumes'], eos.dict()['energies'], c = 'b')
plt.title("Equation of State",size=16)
plt.xlabel(f"Volume ($Ã^{{3}}$)",size=14)
plt.ylabel("Energies (eV/atom)",size=14)
plt.show()
```

```python
eos.dict().keys()
```

```python
eos.dict()
```

```python
with MPRester(mp_api_key) as mpr:
    phonon = mpr.materials.phonon.get_data_by_id('mp-149').ph_bs
    
# make sure you install pymatgen, phonopy
from pymatgen.phonon.plotter import PhononDosPlotter, PhononBSPlotter

plotter = PhononBSPlotter(phonon)
plotter.show()
```

```python
# One can find the available fields this way and include in the fields while querying #

with MPRester(mp_api_key) as mpr:
    print(mpr.summary.available_fields)
```

#### Conditional Querying

```python
with MPRester(mp_api_key) as mpr:
    den_range_cubic = mpr.materials.search(crystal_system="Cubic", 
                                           density=(3,3.5), 
                                           num_elements = 2,
                                           elements = ['C'],
                                           fields = ['density', 'formula_pretty', 'material_id']
                                          )
den_range_cubic[0].dict()
```

```python
with MPRester(mp_api_key) as mpr:
    thermo_nrg_range = mpr.thermo.search(is_stable=True, num_elements = (6,7), fields = ['energy_per_atom', 'symmetry'])
```

# Jarvis

[JARVIS](https://jarvis.nist.gov/), which stands for "Joint Automated Repository for Various Integrated Simulations," is a materials informatics platform developed by NIST (National Institute of Standards and Technology). It's a real-world project aimed at accelerating materials discovery and development through advanced data analytics and machine learning. And it is similar in spirit to the materials project database.

In JARVIS, we have access to precompiled datasets and one could access the [installation page](https://pages.nist.gov/jarvis/) but the following should do

    pip install -U jarvis-tools

Available datasets can be looked up [here](https://pages.nist.gov/jarvis/databases/).

```python
from jarvis.db.figshare import data

x = data('vacancydb') # Use the dataset name as in the link provided for different data #
df_vacdb = pd.DataFrame(x)
```
